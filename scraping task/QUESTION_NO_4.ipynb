```python
import re
import pandas as pd

data = []

for i, fst_pro in enumerate(products):

    # =========================
    # FIRST 5 ROWS SKIP
    # =========================
    if i < 5:
        continue

    dict_pro = {
        "company_name": "",
        "member_name": "",
        "phone": "",
        "websites": "",
    }

    # =========================
    # COMPANY NAME
    # =========================
    comp = fst_pro.find("span", class_="headline")

    if comp:
        dict_pro["company_name"] = comp.get_text(strip=True)

    # =========================
    # DETAILS
    # =========================
    sub = fst_pro.find("div", class_="btSubTitle")

    if sub:

        lines = [x.strip() for x in sub.get_text("\n").split("\n") if x.strip()]

        # =========================
        # PHONE NUMBER
        # =========================
        phone_list = []

        for line in lines:

            line_l = line.lower()

            if (
                "http" in line_l
                or "www" in line_l
                or ".com" in line_l
                or ".pk" in line_l
                or "@" in line_l
            ):
                continue

            if re.search(r"(\+92|0\d{2,4}|111\-)", line):

                clean_phone = line.replace("Tel:", "").strip()

                phone_list.append(" " + clean_phone)

        dict_pro["phone"] = " | ".join(phone_list)

        # =========================
        # WEBSITE
        # =========================
        for line in lines:

            line_l = line.lower()

            if (
                ("http" in line_l or "www" in line_l or ".com" in line_l or ".pk" in line_l)
                and "@" not in line_l
            ):
                dict_pro["websites"] = line.strip()
                break

        # =========================
        # MEMBER NAME
        # =========================
        for line in lines:

            line_l = line.lower()

            if (
                "member" not in line_l
                and not re.search(r"(\+92|0\d{2,4}|111\-)", line_l)
                and "http" not in line_l
                and "www" not in line_l
                and "@" not in line_l
                and ".com" not in line_l
                and ".pk" not in line_l
            ):
                dict_pro["member_name"] = line.strip()
                break

    data.append(dict_pro)

# =========================
# DATAFRAME
# =========================
df = pd.DataFrame(data)

# =========================
# SAVE CSV
# =========================
df.to_csv("product34.csv", index=False, encoding="utf-8-sig")

print(df.head())
```
